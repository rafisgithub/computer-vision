```python
from google.colab.patches import cv2_imshow
import cv2
# reading image from google drive
image = cv2.imread("/content/drive/MyDrive/data/rafi.jpg")
# print(image)
cv2_imshow(image)
```

```python
# Resize the image
resize_image = cv2.resize(image, (200, 200))
cv2_imshow(resize_image)
```

```python
# gray scale image

gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_image)
```

```python
# hsv image

hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
cv2_imshow(hsv_image)
```

```python
# shape of the image

print(image.shape)
```

```python
# blur the image

blur_image = cv2.blur(image, (10, 2))
cv2_imshow(blur_image)
```

```python
# crop the image

crop_image = image[100:500, 200:460]
cv2_imshow(crop_image)
```
