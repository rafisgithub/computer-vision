```python
from google.colab.patches import cv2_imshow

import cv2
image = cv2.imread("/content/drive/MyDrive/data/rafi.jpg")


# print(image)
cv2_imshow(image)
```

```python
#Image flip
flip_image = cv2.flip(image, 1) # 0 mean vertical 1 horizontal and -1 mean both
cv2_imshow(flip_image)
```

```python
#edge detection

image_canny = cv2.Canny(image, 100,300)
cv2_imshow(image_canny)
```

```python
gray_scale_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_scale_image)
# print(gray_scale_image.shape)
```

```python
#historgram
import matplotlib.pyplot as plt

gray_scale_image = cv2.resize(gray_scale_image, (300,300))
cv2_imshow(gray_scale_image)

histogram_image = cv2.calcHist([gray_scale_image], [0], None, [256], [0,256]) # none mean no kernal and [256] holo color group
plt.xlim([0,255])
plt.plot(histogram_image)
plt.xlabel("indensity")
plt.ylabel("frequency")
plt.show()

```
